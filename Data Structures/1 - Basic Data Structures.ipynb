## Singly Linked List

```python
class Node():
    
    def __init__(self, key):
        self.key = key
        self.next = None

class SinglyLinkedList():
    
    def __init__(self):
        self.head = None
        self.tail = None
        
    def push_front(self, key):
        node = Node(key)
        node.next = self.head
        self.head = node
        if self.tail == None:
            self.tail = self.head
    
    def pick_front(self):
        if self.head == None:
            print('Error! Linked list is empty')
        else:
            return self.head.key
    
    def pop_front(self):
        if self.head == None:
            print('Error! Linked list is empty')
        else:
            self.head = self.head.next
        if self.head == None:
            self.tail = self.head
    
    def push_back(self, key):
        node = Node(key)
        node.next = None
        if self.tail == None:
            self.tail = node
            self.head = self.tail
        else:
            self.tail.next = node
            self.tail = self.tail.next
        
    def pick_back(self):
        if self.tail == None:
            print('Error! Linked list is empty')
        else:
            return self.tail.key
        
    def pop_back(self):
        if self.tail == None:
            print('Error! Linked list is empty')
        elif self.head == self.tail:
            self.tail = None
            self.head = None
        else:
            temp = self.head
            while(temp.next.next != None):
                temp = temp.next
            temp.next = None
            self.tail = temp
            
    def is_empty(self):
        return self.head == None
    
    def find_key(self, key):
        temp = self.head
        while(temp != None and temp.key != key):
            temp = temp.next
        if self.head == None:
            print('Linked list is empty')
        elif temp == None:
            print('Not in the linked list')
        else:
            return temp
             
    def insert_after(self, node, key):
        address = self.head
        while(address != None and address.key != node):
            address = address.next
        if self.head == None:
            print('Linked list is empty')
        elif address == None:
            print('Not in the linked list')
        else:  
            new_node = Node(key)
            new_node.next = address.next
            address.next = new_node
            if new_node.next == None:
                self.tail = new_node
        
    def insert_before(self, node, key):
        address = self.head
        if address == None:
            print('Linked list is empty')
        elif address.key == node:
            self.push_front(key)
        else:
            while(address.next != None and address.next.key != node):
                address = address.next
            if address.next == None and address.key != node:
                print('Not in the linked list')
            else:
                new_node = Node(key)
                new_node.next = address.next
                address.next = new_node

    def delete_key(self, key):
        temp = self.head
        if temp == None:
            print("Linked list is empty")
        elif temp.key == key:
            self.pop_front()
        else:
            while(temp.next != None and temp.next.key != key):
                temp = temp.next
            if temp.next == None and temp.key != key:
                print('Not in the linked list')
            else:
                if temp.next.next != None:
                    temp.next = temp.next.next
                else:
                    self.tail = temp
                    temp.next = None
    
    def print_linked_list(self):
        temp = self.head
        if temp == None:
            print("Linked list is empty")
        while(temp != None):
            if temp != self.tail:
                print(temp.key, "-->", end = ' ')
            else:
                print(temp.key, end = ' ')
            temp = temp.next
        
if __name__ == '__main__':
    my_linked_list = SinglyLinkedList()
    my_linked_list.push_front(7)
    my_linked_list.print_linked_list()
    print('\n')
    my_linked_list.push_front(5)
    my_linked_list.print_linked_list()
    print('\n')
    my_linked_list.push_back(6)
    my_linked_list.print_linked_list()
    print('\n')
    my_linked_list.push_back(2)
    my_linked_list.print_linked_list()
    print('\n')
    my_linked_list.insert_after(2, 222)
    my_linked_list.print_linked_list()
    print('\n')
    my_linked_list.insert_before(5, 555)
    my_linked_list.print_linked_list()
    print('\n')
    my_linked_list.pop_back()
    my_linked_list.print_linked_list()
    print('\n')
    my_linked_list.pop_front()
    my_linked_list.print_linked_list()
    print('\n')
    my_linked_list.insert_after(7, 777)
    my_linked_list.print_linked_list()
    print('\n')
    my_linked_list.delete_key(777)
    my_linked_list.print_linked_list()
    print('\n')
```

## Doubly Linked List

```python
class Node():
    
    def __init__(self, key):
        self.key = key
        self.next = None
        self.prev = None

class DoublyLinkedList():
    
    def __init__(self):
        self.head = None
        self.tail = None
        
    def push_front(self, key):
        node = Node(key)
        node.next = self.head
        node.prev = None
        self.head = node
        if node.next != None:
            node.next.prev = node
        if self.tail == None:
            self.tail = self.head
    
    def pick_front(self):
        if self.head == None:
            print('Error! Linked list is empty')
        else:
            return self.head.key
    
    def pop_front(self):
        if self.head == None:
            print('Error! Linked list is empty')
        else:
            self.head = self.head.next
            if self.head != None:
                self.head.prev = None
        if self.head == None:
            self.tail = self.head
    
    def push_back(self, key):
        node = Node(key)
        node.next = None
        node.prev = self.tail
        if self.tail == None:
            self.tail = node
            self.head = node
        else:
            self.tail.next = node
            self.tail = self.tail.next
        
    def pick_back(self):
        if self.tail == None:
            print('Error! Linked list is empty')
        else:
            return self.tail.key
        
    def pop_back(self):
        if self.tail == None:
            print('Error! Linked list is empty')
        elif self.head == self.tail:
            self.tail = None
            self.head = None
        else:
            self.tail = self.tail.prev
            self.tail.next = None
            
    def is_empty(self):
        return self.head == None
    
    def find_key(self, key):
        temp = self.head
        while(temp != None and temp.key != key):
            temp = temp.next
        if self.head == None:
            print('Linked list is empty')
        elif temp == None:
            print('Not in the linked list')
        else:
            return temp
             
    def insert_after(self, node, key):
        address = self.head
        while(address != None and address.key != node):
            address = address.next
        if self.head == None:
            print('Linked list is empty')
        elif address == None:
            print('Not in the linked list')
        else:   
            new_node = Node(key)
            new_node.next = address.next
            new_node.prev = address
            address.next = new_node
            if new_node.next != None:
                new_node.next.prev = new_node
            if new_node.next == None:
                self.tail = new_node
        
    def insert_before(self, node, key):
        address = self.tail
        while(address != None and address.key != node):
            address = address.prev
        if self.tail == None:
            print('Linked list is empty')
        elif address == None:
            print('Not in the linked list')
        else:   
            new_node = Node(key)
            new_node.prev = address.prev
            new_node.next = address
            address.prev = new_node
            if new_node.prev != None:
                new_node.prev.next = new_node
            if new_node.prev == None:
                self.head = new_node

    def delete_key(self, key):
        temp = self.head
        while(temp != None and temp.key != key):
            temp = temp.next
        if self.head == None:
            print('Linked list is empty')
        elif temp == None:
            print('Not in the linked list')
        else:
            if temp.prev != None:
                temp.prev.next = temp.next
            else:
                self.head = temp.next
            if temp.next != None:
                temp.next.prev = temp.prev
            else:
                self.tail = temp.prev
    
    def print_linked_list(self):
        temp = self.head
        if temp == None:
            print("Linked list is empty")
        while(temp != None):
            if temp != self.tail:
                print(temp.key, "<-->", end = ' ')
            else:
                print(temp.key, end = ' ')
            temp = temp.next

if __name__ == '__main__':
    my_linked_list = DoublyLinkedList()
    my_linked_list.push_front(8)
    my_linked_list.print_linked_list()
    print('\n')
    my_linked_list.push_front(1)
    my_linked_list.print_linked_list()
    print('\n')
    my_linked_list.push_back(9)
    my_linked_list.print_linked_list()
    print('\n')
    my_linked_list.push_back(4)
    my_linked_list.print_linked_list()
    print('\n')
    my_linked_list.insert_after(2, 222)
    my_linked_list.print_linked_list()
    print('\n')
    my_linked_list.insert_before(9, 999)
    my_linked_list.print_linked_list()
    print('\n')
    my_linked_list.pop_back()
    my_linked_list.print_linked_list()
    print('\n')
    my_linked_list.pop_front()
    my_linked_list.print_linked_list()
    print('\n')
    my_linked_list.delete_key(777)
    my_linked_list.print_linked_list()
    print('\n')
    my_linked_list.delete_key(8)
    my_linked_list.print_linked_list()
    print('\n')
```

## Stack (Implementation with Array)

```python
class Stack():
    
    def __init__(self):
        self.stack = []
        
    def push(self, key):
        self.stack.append(key)
        return self.stack

    def pick(self):
        if len(self.stack) != 0:
            return self.stack[-1]
        else:
            return "Error! The stack is empty. No element in the stack to show."

    def pop(self):
        if len(self.stack) != 0:
            self.stack.pop()
            return self.stack
        else:
            return "Error! The stack is empty. No element in the stack to remove."
                      
    def empty(self):
        return len(self.stack) == 0

if __name__ == '__main__':
    stack = Stack()
    print(stack.push(3))
    print(stack.push(6))
    print(stack.push(1))
    print(stack.pop())
    print(stack.push(8))
```

## Queue (Implementation with Array)

```python
class Queue():
    
    def __init__(self):
        self.queue = []
        
    def enqueue(self, key):
        self.queue.append(key)
        return self.queue

    def pick(self):
        if len(self.queue) != 0:
            return self.queue[0]
        else:
            return "Error! The queue is empty. No element in the queue to show."

    def dequeue(self):
        if len(self.queue) != 0:
            self.queue.pop(0)
            return self.queue
        else:
            return "Error! The queue is empty. No element in the queue to remove."
                      
    def empty(self):
        return len(self.queue) == 0

if __name__ == '__main__':
    stack = Queue()
    print(stack.enqueue(3))
    print(stack.enqueue(6))
    print(stack.enqueue(1))
    print(stack.dequeue())
    print(stack.enqueue(8))
```

### Check Brackets in the Code

**Task:** Your friend is making a text editor for programmers. He is currently working on a feature that will
find errors in the usage of different types of brackets. Code can contain any brackets from the set
[]{}(), where the opening brackets are [,{, and ( and the closing brackets corresponding to them
are ],}, and ).

For convenience, the text editor should not only inform the user that there is an error in the usage
of brackets, but also point to the exact place in the code with the problematic bracket. First priority
is to find the first unmatched closing bracket which either doesn’t have an opening bracket before it,
like ] in ](), or closes the wrong opening bracket, like } in ()[}. If there are no such mistakes, then
it should find the first unmatched opening bracket without the corresponding closing bracket after it,
like ( in {}([]. If there are no mistakes, text editor should inform the user that the usage of brackets
is correct.

Apart from the brackets, code can contain big and small latin letters, digits and punctuation marks.
More formally, all brackets in the code should be divided into pairs of matching brackets, such that in
each pair the opening bracket goes before the closing bracket, and for any two pairs of brackets either
one of them is nested inside another one as in ($foo[bar]$) or they are separate as in $f(a,b)-g[c]$.
The bracket [ corresponds to the bracket ], { corresponds to }, and ( corresponds to ).

**Input Format:** Input contains one string $S$ which consists of big and small latin letters, digits, punctuation
marks and brackets from the set []{}().

**Constraints:** The length of $S$ is at least $1$ and at most $10^5$.

**Output Format:** If the code in $S$ uses brackets correctly, output **“Success"** (without the quotes). Otherwise,
output the **1-based index of the first unmatched closing bracket**, and if there are no unmatched closing
brackets, output the **1-based index of the first unmatched opening bracket**.

```python
def isBalanced(s):
    stack = []
    for i in range(len(s)):
        if s[i] in ['[','{','(']:
            stack.append(s[i])
            stack.append(i)
            #print('stack', stack)
        elif s[i] not in [')','}',']']:
            continue
        elif len(stack) == 0:
            #print(f'closing bracket {s[i]} at index {i}')
            return i
        else:
            index = stack.pop()
            x = stack.pop()
            if (x == '(' and s[i] != ')') or (x == '{' and s[i] != '}') or (x == '[' and s[i] != ']'):
                #print(f'closing bracket {s[i]} at index {i} dismatches with opening bracket {x} at index {index}')
                return i
            #print('match:', x, s[i])
            #print('stack', stack)
    if len(stack) == 0:
        return "Success"
    else:
        #print(f'opening bracket {stack[0]} at index {stack[1]}')
        return stack.pop()
```

### Compute tree height

**Task:** You are given a description of a rooted tree. Your task is to compute and output its height. Recall
that the height of a (rooted) tree is the maximum depth of a node, or the maximum distance from a
leaf to the root. You are given an arbitrary tree, not necessarily a binary tree.

**Input Format:** The first line contains the number of nodes $n$. The second line contains $n$ integer numbers
from $−1$ to $n − 1$ — parents of nodes. If the $i$-th one of them $(0 \leq i \leq n − 1)$ is $−1$, node $i$ is the root,
otherwise it’s **0-based index of the parent of $i$-th node**. It is guaranteed that **there is exactly one root**.
It is guaranteed that **the input represents a tree**.

**Constraints:** $1 \leq n \leq 10^5$.

**Output Format:** Output the height of the tree.

```python
import sys
import threading
import time

def compute_height_naive(n, parents):
    max_height = 0
    for vertex in range(n):
        height = 0
        current = vertex
        while current != -1:
            height += 1
            current = parents[current]
        max_height = max(max_height, height)
    return max_height

def compute_height(n, parents):
    assert n == len(parents)
    assert parents.count(-1) == 1
    global maximum, count, back
    father = n - 1
    back = n - 1
    maximum = 0
    count = 0
    height(parents, father)
    return maximum
    
def height(parents, father): 
    global maximum, count, back
    if father == -1:
        back = back - 1
        father = back
        count = 0
        if back != -1:
            height(parents, father)
        return maximum
    else:
        father = parents[father]
        count += 1
        maximum = max(count, maximum)
        height(parents, father)

def main():
    n = int(input())
    parents = list(map(int, input().split()))
    print(compute_height(n, parents))

sys.setrecursionlimit(10**7)
threading.stack_size(2**27)
threading.Thread(target=main).start()
time.sleep(10)
```

Below, I have improved my code by storing data in a list to avoid repeated recursion for the same node of the tree.

```python
import sys
import threading
import time
sys.setrecursionlimit(10**7)
threading.stack_size(2**27)

class tree_height():
    
    def __init__(self):
        self.n = 0
        self.parents = []
        self.generation = []

    def read_input(self):
        self.n = int(input())
        self.parents = list(map(int, input().split()))
        self.generation = [0] * self.n
        
    def recursion(self, child):
        if self.parents[child] == -1:
            return 1
        if self.generation[child] != 0:
            return self.generation[child]
        else:
            self.generation[child] = 1 + self.recursion(self.parents[child])
            return self.generation[child]
        
    def compute_height(self):
        
        return max([self.recursion(i) for i in range(self.n)])

def main():
    tree = tree_height()
    tree.read_input()
    print(tree.compute_height())
    print(tree.generation)
    
threading.Thread(target=main).start()
time.sleep(10)
```
