### Singly Linked List

```python
class Node():
    
    def __init__(self, key):
        self.key = key
        self.next = None

class SinglyLinkedList():
    
    def __init__(self):
        self.head = None
        self.tail = None
        
    def push_front(self, key):
        node = Node(key)
        node.next = self.head
        self.head = node
        if self.tail == None:
            self.tail = self.head
    
    def pick_front(self):
        if self.head == None:
            print('Error! Linked list is empty')
        else:
            return self.head.key
    
    def pop_front(self):
        if self.head == None:
            print('Error! Linked list is empty')
        else:
            self.head = self.head.next
        if self.head == None:
            self.tail = self.head
    
    def push_back(self, key):
        node = Node(key)
        node.next = None
        if self.tail == None:
            self.tail = node
            self.head = self.tail
        else:
            self.tail.next = node
            self.tail = self.tail.next
        
    def pick_back(self):
        if self.tail == None:
            print('Error! Linked list is empty')
        else:
            return self.tail.key
        
    def pop_back(self):
        if self.tail == None:
            print('Error! Linked list is empty')
        elif self.head == self.tail:
            self.tail = None
            self.head = None
        else:
            temp = self.head
            while(temp.next.next != None):
                temp = temp.next
            temp.next = None
            self.tail = temp
            
    def is_empty(self):
        return self.head == None
    
    def find_key(self, key):
        temp = self.head
        while(temp != None and temp.key != key):
            temp = temp.next
        if self.head == None:
            print('Linked list is empty')
        elif temp == None:
            print('Not in the linked list')
        else:
            return temp
             
    def insert_after(self, node, key):
        address = self.head
        while(address != None and address.key != node):
            address = address.next
        if self.head == None:
            print('Linked list is empty')
        elif address == None:
            print('Not in the linked list')
        else:  
            new_node = Node(key)
            new_node.next = address.next
            address.next = new_node
            if new_node.next == None:
                self.tail = new_node
        
    def insert_before(self, node, key):
        address = self.head
        if address == None:
            print('Linked list is empty')
        elif address.key == node:
            self.push_front(key)
        else:
            while(address.next != None and address.next.key != node):
                address = address.next
            if address.next == None and address.key != node:
                print('Not in the linked list')
            else:
                new_node = Node(key)
                new_node.next = address.next
                address.next = new_node

    def delete_key(self, key):
        temp = self.head
        if temp == None:
            print("Linked list is empty")
        elif temp.key == key:
            self.pop_front()
        else:
            while(temp.next != None and temp.next.key != key):
                temp = temp.next
            if temp.next == None and temp.key != key:
                print('Not in the linked list')
            else:
                if temp.next.next != None:
                    temp.next = temp.next.next
                else:
                    self.tail = temp
                    temp.next = None
    
    def print_linked_list(self):
        temp = self.head
        if temp == None:
            print("Linked list is empty")
        while(temp != None):
            if temp != self.tail:
                print(temp.key, "-->", end = ' ')
            else:
                print(temp.key, end = ' ')
            temp = temp.next
        
if __name__ == '__main__':
    my_linked_list = SinglyLinkedList()
    my_linked_list.push_front(7)
    my_linked_list.print_linked_list()
    print('\n')
    my_linked_list.push_front(5)
    my_linked_list.print_linked_list()
    print('\n')
    my_linked_list.push_back(6)
    my_linked_list.print_linked_list()
    print('\n')
    my_linked_list.push_back(2)
    my_linked_list.print_linked_list()
    print('\n')
    my_linked_list.insert_after(2, 222)
    my_linked_list.print_linked_list()
    print('\n')
    my_linked_list.insert_before(5, 555)
    my_linked_list.print_linked_list()
    print('\n')
    my_linked_list.pop_back()
    my_linked_list.print_linked_list()
    print('\n')
    my_linked_list.pop_front()
    my_linked_list.print_linked_list()
    print('\n')
    my_linked_list.insert_after(7, 777)
    my_linked_list.print_linked_list()
    print('\n')
    my_linked_list.delete_key(777)
    my_linked_list.print_linked_list()
    print('\n')
```

### Doubly Linked List

```python
class Node():
    
    def __init__(self, key):
        self.key = key
        self.next = None
        self.prev = None

class DoublyLinkedList():
    
    def __init__(self):
        self.head = None
        self.tail = None
        
    def push_front(self, key):
        node = Node(key)
        node.next = self.head
        node.prev = None
        self.head = node
        if node.next != None:
            node.next.prev = node
        if self.tail == None:
            self.tail = self.head
    
    def pick_front(self):
        if self.head == None:
            print('Error! Linked list is empty')
        else:
            return self.head.key
    
    def pop_front(self):
        if self.head == None:
            print('Error! Linked list is empty')
        else:
            self.head = self.head.next
            if self.head != None:
                self.head.prev = None
        if self.head == None:
            self.tail = self.head
    
    def push_back(self, key):
        node = Node(key)
        node.next = None
        node.prev = self.tail
        if self.tail == None:
            self.tail = node
            self.head = node
        else:
            self.tail.next = node
            self.tail = self.tail.next
        
    def pick_back(self):
        if self.tail == None:
            print('Error! Linked list is empty')
        else:
            return self.tail.key
        
    def pop_back(self):
        if self.tail == None:
            print('Error! Linked list is empty')
        elif self.head == self.tail:
            self.tail = None
            self.head = None
        else:
            self.tail = self.tail.prev
            self.tail.next = None
            
    def is_empty(self):
        return self.head == None
    
    def find_key(self, key):
        temp = self.head
        while(temp != None and temp.key != key):
            temp = temp.next
        if self.head == None:
            print('Linked list is empty')
        elif temp == None:
            print('Not in the linked list')
        else:
            return temp
             
    def insert_after(self, node, key):
        address = self.head
        while(address != None and address.key != node):
            address = address.next
        if self.head == None:
            print('Linked list is empty')
        elif address == None:
            print('Not in the linked list')
        else:   
            new_node = Node(key)
            new_node.next = address.next
            new_node.prev = address
            address.next = new_node
            if new_node.next != None:
                new_node.next.prev = new_node
            if new_node.next == None:
                self.tail = new_node
        
    def insert_before(self, node, key):
        address = self.tail
        while(address != None and address.key != node):
            address = address.prev
        if self.tail == None:
            print('Linked list is empty')
        elif address == None:
            print('Not in the linked list')
        else:   
            new_node = Node(key)
            new_node.prev = address.prev
            new_node.next = address
            address.prev = new_node
            if new_node.prev != None:
                new_node.prev.next = new_node
            if new_node.prev == None:
                self.head = new_node

    def delete_key(self, key):
        temp = self.head
        while(temp != None and temp.key != key):
            temp = temp.next
        if self.head == None:
            print('Linked list is empty')
        elif temp == None:
            print('Not in the linked list')
        else:
            if temp.prev != None:
                temp.prev.next = temp.next
            else:
                self.head = temp.next
            if temp.next != None:
                temp.next.prev = temp.prev
            else:
                self.tail = temp.prev
    
    def print_linked_list(self):
        temp = self.head
        if temp == None:
            print("Linked list is empty")
        while(temp != None):
            if temp != self.tail:
                print(temp.key, "<-->", end = ' ')
            else:
                print(temp.key, end = ' ')
            temp = temp.next

if __name__ == '__main__':
    my_linked_list = DoublyLinkedList()
    my_linked_list.push_front(8)
    my_linked_list.print_linked_list()
    print('\n')
    my_linked_list.push_front(1)
    my_linked_list.print_linked_list()
    print('\n')
    my_linked_list.push_back(9)
    my_linked_list.print_linked_list()
    print('\n')
    my_linked_list.push_back(4)
    my_linked_list.print_linked_list()
    print('\n')
    my_linked_list.insert_after(2, 222)
    my_linked_list.print_linked_list()
    print('\n')
    my_linked_list.insert_before(9, 999)
    my_linked_list.print_linked_list()
    print('\n')
    my_linked_list.pop_back()
    my_linked_list.print_linked_list()
    print('\n')
    my_linked_list.pop_front()
    my_linked_list.print_linked_list()
    print('\n')
    my_linked_list.delete_key(777)
    my_linked_list.print_linked_list()
    print('\n')
    my_linked_list.delete_key(8)
    my_linked_list.print_linked_list()
    print('\n')
```

### Stack (Implementation with Array)

```python
class Stack():
    
    def __init__(self):
        self.stack = []
        
    def push(self, key):
        self.stack.append(key)
        return self.stack

    def pick(self):
        if len(self.stack) != 0:
            return self.stack[-1]
        else:
            return "Error! The stack is empty. No element in the stack to show."

    def pop(self):
        if len(self.stack) != 0:
            self.stack.pop()
            return self.stack
        else:
            return "Error! The stack is empty. No element in the stack to remove."
                      
    def empty(self):
        return len(self.stack) == 0

if __name__ == '__main__':
    stack = Stack()
    print(stack.push(3))
    print(stack.push(6))
    print(stack.push(1))
    print(stack.pop())
    print(stack.push(8))
```

### Queue (Implementation with Array)

```python
class Queue():
    
    def __init__(self):
        self.queue = []
        
    def enqueue(self, key):
        self.queue.append(key)
        return self.queue

    def pick(self):
        if len(self.queue) != 0:
            return self.queue[0]
        else:
            return "Error! The queue is empty. No element in the queue to show."

    def dequeue(self):
        if len(self.queue) != 0:
            self.queue.pop(0)
            return self.queue
        else:
            return "Error! The queue is empty. No element in the queue to remove."
                      
    def empty(self):
        return len(self.queue) == 0

if __name__ == '__main__':
    stack = Queue()
    print(stack.enqueue(3))
    print(stack.enqueue(6))
    print(stack.enqueue(1))
    print(stack.dequeue())
    print(stack.enqueue(8))
```
